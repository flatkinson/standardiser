## The Salts Dictionary

This is derived from the ChEMBL salt dictionary, and is simply a list of (neutral) SMILES. Recall that bonds to metals are broken, rule-based standardization and charge neutraliszation are all performed before salt matching. Currently, in order to maximise coverage, stereochemistry is removed and matching done on InChIs.

Note that *any* fragment containing a 'non-organic' element (_i.e._ anything not H, C, N, O, S, P, F, Cl, Br or I) is treated as a salt, so some of the entries here are redundant.

```python
import pandas
from rdkit.Chem import PandasTools

salts = pandas.DataFrame(unsalt.salts.values())

salts.sort("name", inplace=True)

PandasTools.AddMoleculeColumnToFrame(salts, smilesCol='SMILES', molCol='molecule')

salts
```

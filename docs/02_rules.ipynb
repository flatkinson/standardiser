# **`rules`**: apply structure-normalisation transformations

### Introduction

This document describes the **`rules`** module, which encodes a set of rules that are applied in order to standardise molecular representation.

Recall that bonds to Group I & II metals are broken and a round of protonation/deprotonation-based neutralisation are done _before_ these rules are applied, and that a further round of neutralisation is carried out afterwards.

The neutralisation step before rule application is particularly worthy of note, as the it can mean some rules (_i.e._ those for charge-seperated systems) are apparently redundant in some situations (_e.g._ in unsubsituted analogues). The rules are kept as they can still apply in some situations (_e.g._ in fully subsituted analogues) and/or in case the rules module is used stand-alone, where the charge-neutralisation rules may still be useful. If this redundancy is confusing, the rules file could be pruned.

Also note that charge-neutralisation rules _could_ provide an alternative to the proton-addition/removal strategy employed by the neutralise module; however, the latter is used here as it allows finer control, which is particularly useful when more complex systems such as zwitterions are encountered.

I would be interested in missing rules, and in cases where rules need to be restricted (_i.e._ they are being applied inappropriately). I have attempted to highlight some cases where this might be an issue, in order to prompt discussion.

Finally, note that the product is not claimed to be a canonical tautomer: there may be cases where different input tautomers will produce different outputs, and it is up to the user to track this if it is important.

## <a name="Charge-seperated_systems" />Charge-seperated systems

The following rules address cases in which positive and negative formal charges are in conjugation, and the molecule can be neutralised _via_ successive rearrangement of adjacent double and single bonds.

Undesirable interaction with the protonation/deprotonation-based neutralisation steps might conceivably occur, but most likely this won't happen often due to the 'zwitterionic' character of the charge-seperated systems: see the documentation for the [neutralise](3_neutralise.ipynb) module.

In some cases, such as for the rule '[Fix 1,3 charge-seperated systems (aromatic 2)](#Fix_1_3_charge-seperated_systems_aromatic_2)', it is not clear  whether the uncharged imine produced is more or less desirable that the original zwitterion or even than the cationic methyl-pyridine that would be produced by protonation of the amide anion.

This situation is similar to the problem with some of the rules described in the section '[Conjugated cations](#Conjugated_cations)' and, again, caution should be applied in using some of these rules (which is noted below).

<a name="Fix_1_3_charge-seperated_systems_non-aromatic" />
### Fix 1,3 charge-seperated systems (non-aromatic)

```python
demo("C[N-]C(C)=[N+](C)C")
```

Note that this rule (and similar ones below) is illustrated using the fully-substituted moiety: in simpler un- or partially-substituted systems, the **`neutralise`** module will typically handle these systems...

```python
neutralise.apply(s2m("[NH-]C(C)=[NH2+]"))
```

This could conceivably lead to problems in aromatic systems where potentially undesirable imine species might be produced, although this doesn't seem to be a problem in practice as the undesirable species are fixed by application of the rules. This is  illustrated for the rule '[Fix 1,3 charge-seperated systems (aromatic 1)](#Fix_1_3_charge-seperated_systems_aromatic_1)' below.

In some cases the **`neutralise`** module may _not_ charge balance these systems if they appear to be potential zwitterion (_i.e._ where the positively-charged nitrogen is quaternary); see the [documentation](3_neutralise.ipynb) for that module for futher details.

Note that cases where the terminal negatively-charged atom is part of _e.g._ an N-oxide will not now be touched. This modification was motivated by molecules such as [CHEMBL8248](https://www.ebi.ac.uk/chembl/compound/inspect/CHEMBL8248), where application of the basic rule was undesirable...

```python
rules.apply(s2m(i2s("CHEMBL8248")))
```

<a name="Fix_1_3_charge-seperated_systems_aromatic_1" />
### Fix 1,3 charge-seperated systems (aromatic 1)

```python
demo("[n-]1c(=[N+](C)C)cccc1")
```

Note that in the un- or partially-substituted case, when using the full standardise module, the initial application of the **neutralise** module can remove the charges _via_ protonation/deprotonation to give a potentially undesirable hydropyridine-imine...

```python
neutralise.apply(s2m("[n-]1c(=[NH+]C)cccc1"))
```

However, this is then fixed by application of the appropriate 'hydropyridine-imine -> aminopyridine' transform during the rule-application step, so the desired parent is actually obtained...

```python
standardise.apply(s2m("[n-]1c(=[NH+]C)cccc1"))
```

<a name="Fix_1_3_charge-seperated_systems_aromatic_2" />

## <a name="Conjugated_cations" />Conjugated cations

This set of rules affects molecules where a non-protonated atom bearing a positive charge (most commonly a quaternary nitrogen) is conjugated with a neutral nitrogen bearing a hydrogen atom.

In such cases, the charge may be moved from the 'quat' atom to the proton-bearing atom _via_ successive rearrangement of adjacent souble and single bonds. The =(NH+)- moiety thus exposed may then be deprotonated in a subsequent neutralisation step, leaving the neutral parent species.


For some rukes, where (six-membered) aromatic rings are involved, the products are iminium-containing species, which, while still formally aromatic under RDKit's definition, are distinctly less so than the original molecule. An example of this is the rule '[Fix 1,3 conjugated cation (aromatic 2)](#Fix_1_3_conjugated_cation_aromatic_2)' below. It is arguable which species is preferable, but it may well be the more aromatic cation. This issue perhaps isn't so serious with five-membered heterocycles, where the 'loss of aromaticity' doesn't appear so problematic.



This situation is similar to the problem with some of the rules described in the section '[Charge-seperated systems](#Charge-seperated_systems)' and, again, caution should be applied in using some of these rules (which is noted below).

<a name="Fix_1_3_conjugated_cation_non-aromatic" />
### Fix 1,3 conjugated cation (non-aromatic)

<a name="Charge-seperate_sulphoxides" />
### Charge-seperate sulphoxides

The motivation for this rule is that the charge-seperated representation is consistent with that for five-valent N (which is actually applied by default by RDKit on building such molecules) and also that it better reflects the sulphur's 'tetrahedral' (and possibly chiral) geometry. Note that, by contrast, the default representation for a sulphone in ChEMBL is in the hypervalent form, i.e. R-S(=O)(=O)-R.

Question: is this treatment of sulphoxides desirable in the current context? It might well be so in a compound-registration context, but might be inappropriate in a modelling situation.

```python
demo("CS(=O)C")
```

_N.B._ RDKit charge-seperates nitro- and n-oxides _etc._ automatically, thus no rules are needed for these
